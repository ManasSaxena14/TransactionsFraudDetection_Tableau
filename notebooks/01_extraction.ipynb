```python
# necessary imports
import pandas as pd
```

```python
# Define paths to raw CSV files
raw_data_path = '../data/raw/'


def load_dataset(path):
    df = pd.read_csv(f'{raw_data_path}{path}')
    return df

# Load all raw CSV files
account_profiles_df = load_dataset("account_profiles.csv")
fraud_patterns_df = load_dataset("fraud_patterns.csv")
network_edges_df = load_dataset("network_edges.csv")
time_series_stats_df = load_dataset("time_series_stats.csv")
transactions_df = load_dataset("transactions.csv")

print("All raw CSV files loaded.")
```

```python
print('--- account_profiles_df ---\nShape:', account_profiles_df.shape)
print('Columns:', account_profiles_df.columns.tolist())
display(account_profiles_df.head())
print('Info:')
account_profiles_df.info()
print('\nMissing Values:')
display(account_profiles_df.isnull().sum().to_frame(name='missing_count'))
```

```python
print('\n--- fraud_patterns_df ---\nShape:', fraud_patterns_df.shape)
print('Columns:', fraud_patterns_df.columns.tolist())
display(fraud_patterns_df)
print('Info:')
fraud_patterns_df.info()
print('\nMissing Values:')
display(fraud_patterns_df.isnull().sum().to_frame(name='missing_count'))
```

```python
print('\n--- network_edges_df ---\nShape:', network_edges_df.shape)
print('Columns:', network_edges_df.columns.tolist())
display(network_edges_df.head())
print('Info:')
network_edges_df.info()
print('\nMissing Values:')
display(network_edges_df.isnull().sum().to_frame(name='missing_count'))
```
