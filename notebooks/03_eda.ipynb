```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100
```

# Loading Cleaned Datasets

```python
cleaned_data_path = '../data/processed/'


def load_dataset(path):
    df = pd.read_csv(f'{cleaned_data_path}{path}')
    return df

# Load all raw CSV files
account_profiles_df_processed = load_dataset("account_profiles_clean.csv")
fraud_patterns_df_processed = load_dataset("fraud_patterns_clean.csv")
network_edges_df_processed = load_dataset("network_edges_clean.csv")
time_series_stats_df_processed = load_dataset("time_series_stats_clean.csv")
transactions_df_processed = load_dataset("transactions_clean.csv")

print("All cleaned CSV files loaded.")
```

# Fraud Overview

```python
fraud_counts = transactions_df_processed['is_fraud'].value_counts()
fraud_percentage = transactions_df_processed['is_fraud'].value_counts(normalize=True) * 100

print("### Fraud vs. Legitimate Transactions Overview")
print("Total Transactions:", len(transactions_df_processed))
print(f"Legitimate Transactions: {fraud_counts.get(False, 0)} ({fraud_percentage.get(False, 0):.2f}%) ")
print(f"Fraudulent Transactions: {fraud_counts.get(True, 0)} ({fraud_percentage.get(True, 0):.2f}%) ")

plt.figure(figsize=(7, 5))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values, palette='viridis')
plt.title('Distribution of Fraudulent vs. Legitimate Transactions')
plt.xlabel('Is Fraudulent')
plt.ylabel('Number of Transactions')
plt.xticks([0, 1], ['Legitimate (False)', 'Fraudulent (True)'])
plt.show()
```

# Amount Distribution Analysis

```python
print("### Transaction Amount Statistics by Fraud Status")
display(transactions_df_processed.groupby('is_fraud')['amount'].describe())

plt.figure(figsize=(12, 6))
sns.histplot(data=transactions_df_processed, x='amount', hue='is_fraud', kde=True, bins=50, log_scale=True)
plt.title('Distribution of Transaction Amount by Fraud Status (Log Scale)')
plt.xlabel('Transaction Amount (Log Scale)')
plt.ylabel('Count')
plt.legend(title='Is Fraud', labels=['Legitimate', 'Fraudulent'])
plt.show()
```

# Time-Based Fraud Trends

```python
print("### Hourly Fraud Trends")
hourly_fraud = transactions_df_processed.groupby('hour_of_day')['is_fraud'].mean() * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x=hourly_fraud.index, y=hourly_fraud.values, marker='o')
plt.title('Fraud Rate by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Fraud Rate (%)')
plt.xticks(range(0, 24))
plt.show()
```

```python
print('\n### Daily Fraud Trends')
daily_fraud = transactions_df_processed.groupby('transaction_day_name')['is_fraud'].mean() * 100

# Order days of the week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_fraud = daily_fraud.reindex(day_order)

plt.figure(figsize=(10, 6))
sns.barplot(x=daily_fraud.index, y=daily_fraud.values, palette='coolwarm')
plt.title('Fraud Rate by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Fraud Rate (%)')
plt.show()
```

```python
print("### Time Series Stats - Fraud Rate Over Time")

plt.figure(figsize=(15, 7))
sns.lineplot(data=time_series_stats_df_processed, x='hour', y='fraud_rate')
plt.title('Overall Fraud Rate Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Fraud Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Categorical Feature Analysis

```python
print("### Fraud Rate by Merchant Category")
merchant_category_fraud = transactions_df_processed.groupby('merchant_category')['is_fraud'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(14, 7))
sns.barplot(x=merchant_category_fraud.index, y=merchant_category_fraud.values, palette='rocket')
plt.title('Fraud Rate by Merchant Category')
plt.xlabel('Merchant Category')
plt.ylabel('Fraud Rate (%)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
print("### Fraud Rate by Device Type")
device_type_fraud = transactions_df_processed.groupby('device_type')['is_fraud'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
sns.barplot(x=device_type_fraud.index, y=device_type_fraud.values, palette='mako')
plt.title('Fraud Rate by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Fraud Rate (%)')
plt.show()
```

# Account Profiles and Risk Scores

```python
print("### Risk Score Distribution for Fraudsters vs. Non-Fraudsters")

plt.figure(figsize=(10, 6))
sns.histplot(data=account_profiles_df_processed, x='risk_score', hue='is_fraudster', kde=True, bins=30, palette='viridis')
plt.title('Risk Score Distribution by Is Fraudster')
plt.xlabel('Risk Score')
plt.ylabel('Count')
plt.legend(title='Is Fraudster', labels=['Non-Fraudster', 'Fraudster'])
plt.show()

print('\n### Average Risk Score by Fraudster Status')
avg_risk_score = account_profiles_df_processed.groupby('is_fraudster')['risk_score'].mean()
print(avg_risk_score)
```

```python
print("### Fraudster Rate by Account Type")
account_type_fraudster = account_profiles_df_processed.groupby('account_type')['is_fraudster'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
sns.barplot(x=account_type_fraudster.index, y=account_type_fraudster.values, palette='magma')
plt.title('Fraudster Rate by Account Type')
plt.xlabel('Account Type')
plt.ylabel('Fraudster Rate (%)')
plt.show()
```

# Network Analysis Insights

```python
print("### Connections where both accounts are fraudulent")
both_fraud_connections = network_edges_df_processed['both_fraud'].value_counts()
both_fraud_percentage = network_edges_df_processed['both_fraud'].value_counts(normalize=True) * 100

print(f"Total Network Edges: {len(network_edges_df_processed)}")
print(f"Connections where both accounts are fraudulent: {both_fraud_connections.get(True, 0)} ({both_fraud_percentage.get(True, 0):.2f}%) ")
print(f"Connections where at least one account is not fraudulent: {both_fraud_connections.get(False, 0)} ({both_fraud_percentage.get(False, 0):.2f}%) ")

plt.figure(figsize=(7, 5))
sns.barplot(x=both_fraud_connections.index, y=both_fraud_connections.values, palette='viridis')
plt.title('Distribution of Network Edges by Both Accounts Fraudulent Status')
plt.xlabel('Both Accounts Fraudulent')
plt.ylabel('Number of Connections')
plt.xticks([0, 1], ['False', 'True'])
plt.show()
```
